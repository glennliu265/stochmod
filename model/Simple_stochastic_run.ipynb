## Simple Stochastic Model

```python
import numpy as np
import matplotlib.pyplot as plt

import tqdm.notebook as tqdm
```

```python
lbd   = 0 # w/m2/decG
alpha = 1 # w/m2
h     = 50


rho = 1026
cp0 = 3850
dt  = 3600*24*30
T0  = 0


conv = dt/(rho*cp0*h)

t_end = 120000
```

```python
eta   = np.random.normal(0,1,t_end)
T     = np.zeros(t_end)*T0   
```

```python
%%time
for t in np.arange(1,t_end):
    T[t] = T[t-1]*np.exp(-lbd*conv) + alpha*eta[t]*conv
    
```

```python
fig,ax = plt.subplots(1,1)
ax.plot(T)
```
